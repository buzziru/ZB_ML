```python
import numpy as np
import matplotlib.pyplot as plt
```

## 다항식

```python
x = np.linspace(-3, 2, 100)
y = 3 * x**2 + 2
```

```python
import matplotlib as mpl
mpl.style.use('seaborn-v0_8')

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$3x^2 +2$', fontsize=25)
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y1 = 3 * x**2 + 2
y2 = 3 * (x+1)**2 + 2

plt.figure(figsize=(8, 6))
plt.plot(x, y1, lw=2, ls='solid', label='$3x^2 + 2$')
plt.plot(x, y2, ls='-.', label='$3(x+1)^2 + 2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='best', fontsize=15)
plt.show()
```

## 지수함수

```python
x = np.linspace(-2, 2, 100)
a11, a12, a13 = 2, 3, 4
y11, y12, y13 = a11**x, a12**x, a13**x

a21, a22, a23 = 1/2, 1/3, 1/4
y21, y22, y23 = a21**x, a22**x, a23**x
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(x, y11, color='k', label='$2^x$')
ax[0].plot(x, y12, '--', color='k', label='$3^x$')
ax[0].plot(x, y13, ':', color='k', label='$4^x$')
ax[0].legend(fontsize=20)

ax[1].plot(x, y21, color='k', label='$(1/2)^x$')
ax[1].plot(x, y22, '--', color='k', label='$(1/3)^x$')
ax[1].plot(x, y23, ':', color='k', label='$(1/4)^x$')
ax[1].legend(fontsize=20)

plt.show()
```

```python
x = np.linspace(0, 10)

plt.figure(figsize=(6, 6))
plt.plot(x, x**2, '--', color='k', label='$x^2$')
plt.plot(x, 2**x, color='k', label='$2^x$')
plt.legend(loc='center left', fontsize=25)
plt.xlabel('$x$', fontsize=25)
plt.ylabel('$y$', fontsize=25)

plt.show()
```

## 로그함수

```python
def log(x, base):
    return np.log(x)/np.log(base)

x1 = np.linspace(0.0001, 5, 1000)
x2 = np.linspace(0.01, 5, 100)

y11, y12 = log(x1, 10), log(x2, np.e)
y21, y22 = log(x1, 1/10), log(x2, 1/np.e)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 8))
ax[0].plot(x1, y11, label='$log_{10}{x}$')
ax[0].plot(x2, y12, ':', label='$log_{e}{x}$')
ax[0].legend()

ax[1].plot(x1, y21, label='$log_{1/10}{x}$')
ax[1].plot(x2, y22, ':', label='$log_{1/e}{x}$')
ax[1].legend()
```

## 시그모이드
- $\sigma(z) = \frac{1}{1 + e^{-z}}$

```python
z = np.linspace(-10, 10, 100)
sigma = 1 / (1 + np.exp(-z))

plt.figure(figsize=(12, 8))
plt.plot(z, sigma, label='$\\frac{1}{1 + e^{-z}}$')
plt.xlabel('$z$', fontsize=25)
plt.ylabel('$\sigma(z)$', fontsize=25)
plt.legend(fontsize=25)
plt.show()
```

## 벡터
- $x = \begin{pmatrix} x_1 \\ x_2 \end{pmatrix}$ $\,\qquad$  $x = (x_1, x_2)^T$
- 단일변수 스칼라 함수 : $y = f(x)$
- 다중변수 스칼라 함수 : $y = f(\vec{x})$
- 다변수 벡터함수 : $F(\mathbf{x}) = (f_1(\mathbf{x}), f_2(\mathbf{x}), \ldots , f_n(\mathbf{x}))^T$

- 다변수 벡터함수 예제
$S(u, v) = \begin{pmatrix} u \\ v \\ 1 + u^2 + \frac{v}{1 + v^2}\end{pmatrix}$

```python
u = np.linspace(0, 1, 30)
v = np.linspace(0, 1, 30)
U, V = np.meshgrid(u, v) # 카티시안 곱
X = U
Y = V
Z = (1 + U**2) + (v/(1 + V**2))
```

```python
fig = plt.figure(figsize=(7, 7))
ax = plt.axes(projection='3d')

ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.zaxis.set_tick_params(labelsize=12)

ax.set_xlabel('$X$', fontsize=20)
ax.set_ylabel('$Y$', fontsize=20)
ax.set_zlabel('$Z$', fontsize=20)

ax.scatter3D(X, Y, Z, marker= '.', color='gray')
# ax.view_init(elev=30, azim=55) # 그래프가 보이는 각도 조정

plt.show()
```

## 합성함수

```python
x = np.linspace(-4, 4, 100)
y = x**3 - 15*x + 30 # f(x)
z = np.log(y)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(x, y, label='$X^3 - 15x + 30$', color='k')
ax[0].legend(fontsize=10)
ax[1].plot(y, z, label='$log{y}$', color='k')
ax[1].legend(fontsize=10)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(x, z, '--', label='$\log{f(x)}$', color='k')
ax[0].legend(fontsize=10)
ax[1].plot(x, y, label='$x^3 - 15x + 30$', color='k')
ax[1].legend(fontsize=10)

ax_tmp = ax[1].twinx()
ax_tmp.plot(x, z, '--', label='$\log{f(x)}$', color='k')
ax_tmp.legend(loc='lower left', fontsize=10)

plt.show()
```
